```python
# Out of the box models cannot answer questions about your company's proprietery data
# Three ways to utilize llms to summarize, analyze, Q& A etc - prompting, fine tuning and RAG
```

```python
# Easiest way to provide custom data is through Prompting with system messages, few shot prompting, chat history
# Suitable for smaller datasets
# fine tuning means continuing to train a pretrained model like gpt4 with your custom dataset. fine tuning directly interferes with the models
# inner working adjusting the wights of the neural network to better  align with the nuances of the the custom data. Requires lot of efforts
# and resources, needs substantial dataset needs ML/DL expertise
# Third technique is RAG, where we keep large amout on embedded data inside a database and retrieve only portion necessary to answer questions
```

```python
# RAG application consists of 
# Indexing  - Preparing the data and storing it in a specialized database - 1.1 Load custom data froom various formats into standard lang chain
# format, spit the documents into smaller chunks to make sure it fits the LLM context window limit, Embed the text in the document so that
# each chunk has its own vector representation capturing the semantic meaning of the text. Store it in a vector store.
# Retrieval - retrieving only document most relevant to the request based on user query
# Generation - taking the user request with the retrieved response and employing an LLM to produce a context aware response.
# Downsides - increased latency due to retrieving data real time. the quality of the retrieved documents heavily influences models response.
```

```python
# Document loading and splitting - Take a data source - json, pdf, html, scraping content and storing it as a list of Document objects
```

```python
# Langchain supports integrations with many file formats and can load directly from various file storage platforms such as a google drive
# after files have loaded as list of documents. they can be of varied sizes. This can cause downstream issues if they are too large with
# model window limit,token sizes. 
# Split the data into chunks
# During Embedding, it converts text to vector capturing its semantic meaning in a numerical format, can perform similarity searches efficiently
```

```python
# when positioned on a coordinate system, the vectors of words that are often used in same context and are close in semantic meaning often occur
#closer to each other.
# When user questions the chat bot, the first step is that user question is created into a vector representation and it gets mapped on the
# coordinate plane
# How to measure closeness of vectors - using dot product, cosine similarity and eucleadean distance
# Words that are close to each other will have smaller angles between them and a larger cosine and vice versa
# Vector embeddings are over 1000 or more dimensions
```

```python
# We now need a place to store the vector embeddings where we can do efficient semantic similarity searches - Vector databases
# Retrievers are part of RAG change responsible for most relevant and diverse parts of the data. Each chunk should create new and valuable
# information to the table.
# Augmented Geneeration - where user input and retrieved text are combined and fed into the llm
```
